```python
# ==========================================================
# Cricket Performance Analysis using NumPy & Matplotlib
# Author : Viren Nagpal
# Description:
# Analyze player performance across 3 matches using NumPy
# and visualize the results using Matplotlib.
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Player Names
# ----------------------------------------------------------

players = np.array([
    "Viren",
    "Rohit",
    "MS Dhoni",
    "Ruturaj"
])

# ----------------------------------------------------------
# Runs scored in 3 matches
# Rows = Players
# Columns = Matches
# ----------------------------------------------------------

runs = np.array([
    [56, 12, 102],
    [91, 70, 0],
    [30, 53, 18],
    [28, 85, 45]
])

# ----------------------------------------------------------
# Balls faced
# ----------------------------------------------------------

balls = np.array([
    [28, 10, 60],
    [54, 41, 2],
    [23, 25, 12],
    [15, 55, 25]
])

# ==========================================================
# TOTAL RUNS
# ==========================================================

total_runs = np.sum(runs, axis=1)

print("\n========== Total Runs ==========")

for i in range(len(players)):
    print(f"{players[i]} : {total_runs[i]}")

# ==========================================================
# AVERAGE RUNS
# ==========================================================

average_runs = np.mean(runs, axis=1)

print("\n========== Average Runs ==========")

for i in range(len(players)):
    print(f"{players[i]} : {average_runs[i]:.2f}")

# ==========================================================
# HIGHEST & LOWEST SCORE
# ==========================================================

highest_score = np.max(runs)
lowest_score = np.min(runs)

print("\nHighest Individual Score :", highest_score)
print("Lowest Individual Score :", lowest_score)

# Player with highest individual score

row, col = np.unravel_index(np.argmax(runs), runs.shape)

print(
    f"\nHighest Individual Score ({highest_score}) "
    f"was scored by {players[row]} in Match {col + 1}"
)

# ==========================================================
# HIGHEST TOTAL RUN SCORER
# ==========================================================

highest_total_player = players[np.argmax(total_runs)]

print("\nHighest Run Scorer :", highest_total_player)

# ==========================================================
# PLAYERS SCORING 50+ IN MATCH 1
# ==========================================================

print("\nPlayers scoring 50+ in Match 1")

indices = np.where(runs[:, 0] >= 50)[0]

for i in indices:
    print(players[i], "-", runs[i, 0])

# ==========================================================
# PLAYER OUT ON DUCK IN MATCH 3
# ==========================================================

print("\nPlayers out on Duck in Match 3")

duck = np.where(runs[:, 2] == 0)[0]

for i in duck:
    print(players[i])

# ==========================================================
# HALF CENTURIES
# ==========================================================

half_centuries = np.sum(runs >= 50)

print("\nTotal Half Centuries :", half_centuries)

# ==========================================================
# STRIKE RATE
# ==========================================================

strike_rate = (runs / balls) * 100

print("\n========== Strike Rate ==========")

for i in range(len(players)):
    print(players[i], ":", np.round(strike_rate[i], 2))

# ==========================================================
# STRIKE RATE >150 IN MATCH 1
# ==========================================================

print("\nStrike Rate >150 in Match 1")

indices = np.where(strike_rate[:, 0] > 150)[0]

for i in indices:
    print(f"{players[i]} : {strike_rate[i,0]:.2f}")

# ==========================================================
# VISUALIZATION 1
# Total Runs Bar Chart
# ==========================================================

plt.figure(figsize=(8, 5))

plt.bar(
    players,
    total_runs,
    color=["red", "green", "blue", "orange"],
    edgecolor="black"
)

plt.title("Total Runs Scored by Players")
plt.xlabel("Players")
plt.ylabel("Runs")
plt.grid(axis="y")

plt.show()

# ==========================================================
# VISUALIZATION 2
# Contribution Pie Chart
# ==========================================================

plt.figure(figsize=(7, 7))

plt.pie(
    total_runs,
    labels=players,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Contribution of Each Player")

plt.show()

# ==========================================================
# VISUALIZATION 3
# Strike Rate Trend of Viren
# ==========================================================

matches = ["Match 1", "Match 2", "Match 3"]

plt.figure(figsize=(8, 5))

plt.plot(
    matches,
    strike_rate[0],
    marker="o",
    linewidth=2
)

plt.title("Strike Rate of Viren")
plt.xlabel("Matches")
plt.ylabel("Strike Rate")

plt.grid(True)

plt.show()
```
